```python
def mad_libs_game():
    # Fancy title display
    print("\n" + "="*50)
    print("ðŸŽ® ULTIMATE MAD LIBS ADVENTURE ðŸŽ®".center(50))
    print("Made by Sarfraz Ahmad".center(50))
    print("="*50 + "\n")

    print("I'll ask you for some words, and then create a funny story with them.")
    print("Let's get started!\n")

    # Get inputs from the user with colorful prompts
    word_types = [
        ("adjective", "that describes a magical place"),
        ("noun", "something living"),
        ("verb_past_tense", "an exciting action"),
        ("adverb", "how someone might move"),
        ("adjective", "a personality trait"),
        ("noun", "a strange creature"),
        ("noun", "a mysterious object"),
        ("adjective", "how something might look"),
        ("verb", "something you might do with treasure"),
        ("adverb", "how you'd celebrate"),
        ("verb_past_tense", "how a story might end"),
        ("adjective", "how a kingdom might feel")
    ]

    user_inputs = {}
    try:
        for i, (word_type, description) in enumerate(word_types, 1):
            key = f"{word_type}{i if word_type in user_inputs else ''}"
            user_inputs[key] = input(f"[{i}/12] Enter a {word_type} ({description}): ")

        # Create a more elaborate story with the user's inputs
        story = f"""
        âœ¨ THE INCREDIBLE ADVENTURE âœ¨

        Once upon a time, in a {user_inputs['adjective']} kingdom far beyond the clouds,
        there lived a {user_inputs['noun']} who {user_inputs['verb_past_tense']} {user_inputs['adverb']}
        through the enchanted forest every morning.

        The {user_inputs['noun']} was known to be extremely {user_inputs['adjective2']} and owned
        a legendary pet {user_inputs['noun2']} that could speak three different languages.

        One fateful day, while exploring an ancient temple, the {user_inputs['noun']}
        discovered a glowing {user_inputs['noun3']} that appeared incredibly {user_inputs['adjective3']}.
        Legends said that whoever possessed this artifact would gain extraordinary powers!

        Without hesitation, the {user_inputs['noun']} decided to {user_inputs['verb']} {user_inputs['adverb2']}
        with the mysterious object, causing the sky to fill with fireworks and rainbows.

        After that magical day, the {user_inputs['noun']} {user_inputs['verb_past_tense2']} happily ever after
        in the {user_inputs['adjective4']} kingdom, becoming the greatest hero the realm had ever known.

        THE END
        """

        # Display the completed story with some flair
        print("\n" + "â˜…"*60)
        print("HERE'S YOUR EPIC MAD LIBS ADVENTURE:".center(60))
        print("â˜…"*60)
        print(story)
        print("â˜…"*60)
        return True
    except KeyboardInterrupt:
        print("\n\nGame interrupted! Returning to main menu...")
        return False

# Run the game
if __name__ == "__main__":
    try:
        mad_libs_game()

        # Ask if the user wants to play again with improved options
        while True:
            try:
                print("\n" + "-"*40)
                play_again = input("Would you like to embark on another adventure? (yes/no): ").lower()
                if play_again in ["yes", "y", "yeah", "sure"]:
                    print("\nExcellent choice! Let's create another story!\n")
                    if not mad_libs_game():
                        break
                elif play_again in ["no", "n", "nope"]:
                    print("\nâœ¨ Thanks for playing Ultimate Mad Libs! Your creativity was amazing! âœ¨")
                    print("See you on your next adventure!")
                    break
                else:
                    print("Please enter 'yes' or 'no'.")
            except KeyboardInterrupt:
                print("\n\nGame interrupted!")
                choice = input("\nDo you want to exit the game? (yes/no): ").lower()
                if choice in ["yes", "y"]:
                    print("\nâœ¨ Thanks for playing Ultimate Mad Libs! âœ¨")
                    print("See you on your next adventure!")
                    break
    except Exception as e:
        print(f"\nAn unexpected error occurred: {e}")
        print("The game will now exit.")
    finally:
        print("\nGame session ended.")
```
